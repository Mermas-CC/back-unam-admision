```python
import requests

url = "http://localhost:11434/api/generate"

payload = {
    "model": "gemma3:1b",  # nombre de tu modelo en Ollama
    "prompt": "Explica qué es el teorema de Bayes en pocas palabras."
}

response = requests.post(url, json=payload)
for line in response.iter_lines():
    if line:
        print(line.decode())
```
